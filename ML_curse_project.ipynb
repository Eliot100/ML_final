```python
!pip install pydub
!pip install scipy
from scipy.io.wavfile import read
from pydub import AudioSegment
import numpy as np
import pandas as pd
```

```python
data = {      'age': [],
              'word': [],
              'signal': [],
              'rate': [],
              'emotion': [],
        }
df = pd.DataFrame(data=data) 
```

```python
sound = AudioSegment.from_wav('/content/drive/MyDrive/TESS Toronto emotional speech set data/YAF_neutral/YAF_back_neutral.wav')
sound
```

```python
rate, signal = read('/content/drive/MyDrive/TESS Toronto emotional speech set data/YAF_neutral/YAF_back_neutral.wav')
rate
```

```python
audio_segment = AudioSegment(
    signal.tobytes(), 
    frame_rate=rate, #24414
    sample_width=signal.dtype.itemsize, 
    channels=1
)
audio_segment
```

```python
import glob
start_age_index = 62
start_word_index = 71
end_word_index = 5
age_size = 3
for foldr in glob.glob('/content/drive/MyDrive/TESS Toronto emotional speech set data/*'):
  age = foldr[start_age_index:(start_age_index+age_size)]
  emotion = foldr[(start_age_index+age_size+1):]
  emotion_size = len(emotion)
  if (emotion == "Pleasant_surprise"):
    emotion_size = 2
  for file_name in glob.glob(foldr+'/*'):
    signal = AudioSegment.from_file(file_name)
    signal = np.array(signal.get_array_of_samples())
    word = file_name[(start_word_index+emotion_size):-(end_word_index+emotion_size)]
    df = df.append(pd.Series([age, word, signal, rate, emotion], index = df.columns), ignore_index=True)
df
```

```python
df.to_csv("/content/drive/MyDrive/TESS Toronto emotional speech set data/TESS_df.csv")
```
